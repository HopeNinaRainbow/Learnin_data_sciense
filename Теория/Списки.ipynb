#### Список в Python — это упорядоченная изменяемая коллекция объектов произвольных типов.

### Методы используемые для списков:

##### .APPEND() - добавить в имеющийся список дополнительный элемент («приписывается» справа).

##### .CLEAR() - очистить список полностью.

##### .COUNT() - в качестве аргумента принимает значение элемента, количество которого нужно посчитать (a.count("b")).

##### .COPY() - копирование списка (a.copy() == a[:]).

##### .EXTEND() - объединить списки.

##### .REVERSE() - развернуть значения в списке.

##### .SORT() - сортировка списка

```python
#В чём разница между remove, del и pop?
#remove() удаляет первое совпадающее значение:

li = ['a','b','c','d']
li.remove('b')
#li
#['a', 'c', 'd']

#del удаляет элемент по его индексу:

li = ['a','b','c','d']
del li[0]
#li
#['b', 'c', 'd']

#pop() удаляет элемент по индексу и возвращает этот элемент:

li = ['a','b','c','d']
li.pop(2)
#'c'
#li
#['a', 'b', 'd']

```
